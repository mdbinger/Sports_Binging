```python
final_tp_list = []
for item in triple_play_list:
    temp_data = item.split()
    final_tp_list.append(temp_data)

i = 0
complete_names_list = []
for i in range(len(final_tp_list)):
    j = 7
    names_list = []
    while j < len(final_tp_list[i]):
        if bool(re.search(r'\d', final_tp_list[i][j])) == False:
            names_list.append(final_tp_list[i][j])
            j = j + 1
        else:
            break
#         print(j)
            
    complete_names_list.append(names_list)
        
complete_names_list
```
